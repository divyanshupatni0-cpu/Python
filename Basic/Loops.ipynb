```python
#for loop
#Print numbers from 1 to 10 using a for loop.
for i in range(1,11):
    print(i)
```

```python
#Print all even numbers between 1 and 20.
for i in range(2,21,2):
    print(i)
```

```python
#Print the square of each number from 1 to 5.
for i in range(1,6):
    print(i*i)
```

```python
#Given a list fruits = ["apple", "banana", "cherry"], print each fruit in the list.
fruits=["apple","banana","chery"]
for i in fruits:
    print(i)
```

```python
#Calculate the sum of all numbers in the list [1, 2, 3, 4, 5].
sum=0
lst=[1,2,3,4,5]
for i in lst:
    sum+=i
print(sum)
```

```python
#Print numbers from 1 to 20, but print "Fizz" for multiples of 3 and "Buzz" for multiples of 5.
for i in range(1, 21):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")   
    elif i % 3 == 0:
        print("Fizz")       
    elif i % 5 == 0:
        print("Buzz")       
    else:
        print(i)  
   
    
```
