```python
#Write a function to print numbers from 1 to N.
def num (n):
    for i in range(1,n+1):
        print(i)
num(4)
```

```python
def fact(n):
    if n==0:
        return 1
    return n*fact(n-1)
print(fact(5))
 
```

```python
#Write a function that returns the sum of all even numbers in a list.
lst=[1,2,4,5,6,7,8]
def change(lst):
    total=0
    for i in lst:
        if i%2==0:
            total+=i
    return total
print(change(lst))
```

```python
#resverse a list using function
def reverse (lst):
    return list(reversed(lst))
lst=[1,2,3,4,5]
print(reverse(lst))

```
