```python
#1.Divide by Zero
#Write a program that takes two numbers from the user and prints their division.
#Handle the case where the denominator is 0 using exception handling.
try:
    a=int(input("entered a number"))
    b=int(input("entered a number"))
    print(a/b)
except ZeroDivisionError:
    print("no zero divide by zero")
except ValueError:
    print("no int multiply with abc alphbet")
```

```python
#Multiple Exceptions
#Write a program that:
#Takes two numbers from the user.
#Divides the first by the second.
#Converts the result to an integer.
#Handle all possible exceptions (ZeroDivisionError, ValueError, TypeError) separately and print appropriate messages.
try:
    a=int(input("entered a number"))
    b=int(input("entered a number"))
    print(a/b)
except ZeroDivisionError:
    print("not divide by zero")
except ValueError:
    print("no multiply by aphannbet")
except TypeError:
    print("typing error")
```

```python
#A programmer writes a function that processes user input, performs a division, and logs the operation. During execution, different exceptions may occur, such as ValueError for invalid input or ZeroDivisionError for division by zero.
#Describe a scenario where the except block would execute.
#Explain a situation where the finally block would execute even if no exception occurs.
#Discuss how the behavior would change if the return statement is present inside the try block
def division(a, b):
    try:
        result = a / b
        print(result)
        return result  
    except ZeroDivisionError:
        print("Not divisible by zero")  
    except TypeError:
        print("Int cannot divide by alphabet")  
    finally:
        print("All blocks execute")  
print(division(5,3))
print(division(5,0))
print(division('abc',5))
```
